### 함수 구성
- content_crawl(url, section)
- 수집한 데이터를 tuple로 구성 후 반환
    - 행추가를 이용해서 df에 저장
    

```python
def content_crawl(url, section):
    # 기사 페이지 소스 요청 및 반환
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
    res = requests.get(url, headers=headers)
    # requests 객체의 반환된 html 코드는 text 속성에 있음
    html = res.text
    bs_obj = bs4.BeautifulSoup(html, 'html.parser')
    # 전체 기사 카드를 포함하고 있는 영역 추출
    news_ls = bs_obj.find('div', {'id':'ct'})
    # title
    title_text = news_ls.find('h2', {'id':'title_area'})
    title = title_text.text
    # date, writer, content
    info_div = news_ls.find('div', {'class':'media_end_head_info'})
    info = info_div.findAll('div')
    date = info[1].span.text
    writer = news_ls.find('em', {'class' : "media_end_head_journalist_name"}).text
    content = bs_obj.find('div',{'id':'dic_area'}).text.strip().replace('\n', '')
    
    return title, date, writer, content, section  # 튜플로 반환
```

```python
# 함수 테스트
a,b,c,d,e = content_crawl(url, '세계')
```

```python
# 빈 df 생성 후 행 추가하는 방식으로 정리
news_content = pd.DataFrame({'title':[], 'data':[], 'writer':[], 'content':[], 'section':[]})
news_content
```

```python
urls = df_head.기사링크
```

```python
i = 0 # 행 인덱스로 사용하기위해 선언한 변수
for url in urls:
    try:
        title,date,writer,content,section = content_crawl(url, df_head['분류'][i])
        news_content.loc[i] = [title,date,writer,content,section]
    except:
        print('error')
    i += 1
```
